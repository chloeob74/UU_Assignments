### c.
**What is the expected winnings of any gamble in European roulette?**

European roulette has 37 numbers: 0-36. 

If you place a $1 bet on a single number:
- Probabiity of winning: $\frac{1}{37}$
- If you win, you gain $35 net ($36 minus the $1 you bet)
- If you lose, you lose $1

Expected value is:

$$
E[winnings] = \frac{1}{37}(+35)+\frac{36}{37}(-1) = \frac{35 - 36}{37} = \frac{-1}{37}
$$

If you bet $1 on red:
- There are 18 red number, 18 black numbers and 1 green zero
- Probability of winning: $\frac{18}{37}$ 
- Probability of winning: $\frac{19}{37}$

$$
E[winnings] = \frac{18}{37}(+1) + \frac{19}{37}(-1) = \frac{18 - 19}{37} = \frac{-1}{37}
$$

$\frac{-1}{37}$ holds the same for all standard bets because the payouts are chosen that way.

So, a stake of $X:
$$
E[\text{net winnings}] = \frac{-1}{37}X \approx -0.027X
$$

### d.
**Imagine you roll a die, and you record the value you get. But, if you roll a six, you roll again, and add that value. What is the expected value?**

Let Y be the total

$$
E[Y] = \sum_{k=1}^5 \frac{1}{6} \cdot k + \frac{1}{6} \cdot (6 + E[\text{second roll}])
$$

From our previous calculations, we know that
$$
E[\text{second roll}] = 3.5
$$

So,

1. First part (first roll 1-5):
$$
\sum_{k=1}^5 \frac{k}{6} = \frac{1 + 2 + 3 + 4 + 5}{6} = \frac{15}{6} = 2.5
$$

2. Second part (first roll is 6)
$$
\frac{1}{6}(6 + 3.5) = \frac{1}{6} \cdot 9.5 = \frac{9.5}{6}
$$

So,

$$
E[Y] = 2.5 + \frac{9.5}{6} = \frac{30}{12} + \frac{19}{12} = \frac{49}{12} \approx 4.0833
$$


### e.
**Imagine that the process described in the last question continues until you fail to roll a six. What is the expected value of the process? (This can be tricky, you can simulate it to get an answer if you prefer. Hint: The answer is 4.2.)**

Let S be the total sum

**Step 1: Model the number of 6's**

Let N = the number of ^'s rolled before the first non-6

For each roll:
- 6 has probability $\frac{1}{6}$
- non-6 has probability $\frac{5}{6}$

So, N is a geometric distribution with succuess probability $p = \frac{5}{6}$ (success = "roll a non-6, taking values 0, 1, 2, ...)

The formula looks like:

$$
P(N = n) = \left( \frac{1}{6} \right)^n \cdot \frac{5}{6}
$$

For this geometric distribution (starting at 0):

$$
E[N] = \frac{1 - p}{p} = \frac{(1/6)}{(5/6)} = \frac{1}{5} = 0.2
$$

**Step 2: Express the total sum:**

Every time you roll a 6, you add 6. After N sixes, you roll one last time and get a non-6 result.

Let Y be the value of the final non-6 roll. It is conditional on being a non-6.
Y is a uniform on {1, 2, 3, 4, 5}, so

$$
E[Y] = \frac{1 + 2 + 3 + 4 + 5}{5} = 3
$$

The total sum is 

$$
S = 6N + Y
$$

**Step 3: Take expectations**

$$
E[S] = E[6N + Y] = 6E[N] + E[Y]
$$

From before, we know that $E[N] = \frac{1}{5}$ and E[Y] = 3, so

$$
E[S] = 6 \cdot \frac{1}{5} + 3 = \frac{6}{5} + 3 = \frac{6}{5} + \frac{15}{5} = \frac{21}{5} = 4.2
$$

Final answer:
$$
E[S] = \frac{21}{5} = 4.2
$$
